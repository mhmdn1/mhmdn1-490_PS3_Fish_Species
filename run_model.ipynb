```python
# run_model.ipynb

import pandas as pd
from neural_network import BuildDataset, NeuralNetwork

# Load the data
data = pd.read_csv('/content/data.csv')

# Prepare the dataset
X_train, X_test, y_train, y_test = BuildDataset(data)

# Initialize and train the neural network
nn = NeuralNetwork(input_shape=X_train.shape[1])
nn.train(X_train, y_train, epochs=100)

# Evaluate the model
evaluation = nn.evaluate(X_test, y_test)
print("Evaluation Metrics (Loss, Accuracy):", evaluation)

# Make predictions and display them alongside actual labels
predictions = nn.predict(X_test)



```
